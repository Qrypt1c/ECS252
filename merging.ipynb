```python
putinDF = russiaDF = ukraineDF = warDF = zelenskyDF = pd.DataFrame()
```

```python
def merging(term):
    totalDF = pd.DataFrame()
    for i in range(1, 13):
        download_temp = json.load(open("Romanian_Scrape/" + term + "_romania" + str(i) + ".json"))
        totalDF = pd.concat([totalDF, pd.DataFrame(download_temp["data"])], ignore_index=True)
    return totalDF
```

```python
putinDF = merging("putin")
russiaDF = merging("russia")
ukraineDF = merging("ukraine")
warDF = merging("war")
zelenskyDF = merging("zelensky")
```

```python
romaniaDF = pd.DataFrame()
romaniaDF = pd.concat([romaniaDF, russiaDF, ukraineDF, warDF, zelenskyDF], ignore_index=True)
```
